### Problem 3
<span  style="color:green; font-size:16px">Filter the diamonds dataframe for only the top three most frequently occurring values of clarity. Then use matplotlib directly to plot a vertical bar plot of all diamond cuts by the top 3 values of clarity. Make sure each value of clarity has a different color (but same color for each cut) and is labeled. Use a loop to iterate through the different clarity groups.</span>

```python
top_three = diamonds.clarity.value_counts().iloc[:3]
top_three
```

```python
d3 = diamonds[diamonds.clarity.isin(top_three.index)]
```

```python
d3_sum = d3.pivot_table(index='cut', columns='clarity', values='color', aggfunc='count')
d3_sum
```

```python
colors = ['r', 'c', 'g']
for i in range(3):
    x = np.arange(len(d3_sum))
    plt.bar(x + (i * .25)-.125, d3_sum.iloc[:, i], width=.25, color=colors[i], label= d3_sum.columns[i])
    
plt.xticks(x + .25, d3_sum.index, size='10')
plt.title('Cut by Clarity', color='r')
plt.legend()
plt.xlim(-.5, 5)
```

### Problem 4
<span  style="color:green; font-size:16px">We will make clever use of the iterability of a groupby object. For the city of Houston dataset, iterate through every RACE GENDER combination and plot a histogram using matplot of their base salary. Don't plot groups with less than 100 employees. Make the figure larger as well and make the limits of the x axis much smaller to capture the bulk of the data.</span>

```python
coh = pd.read_csv('data/coh_employee.csv')
```

```python
plt.figure(figsize=(12,8))
for name, group in coh.groupby(['RACE', 'GENDER'])['BASE_SALARY']:
    if group.size > 100:
        plt.hist(group.dropna(), alpha=.6, label=name[0] + name[1], normed=True)

plt.xlim(30000,120000)
plt.legend()
```

### Problem 6
<span  style="color:green; font-size:16px">Calculate years of experience for city of houston employees and make a scatter plot between years of experience and base salary. **`EXTRA CREDIT`**: Use [statsmodels](http://statsmodels.sourceforge.net/devel/index.html) to find a regression line and draw it through the points.</span>

```python
coh['HIRE_DATE'] = coh['HIRE_DATE'].astype('datetime64')
```

```python
coh['EXPERIENCE'] = (pd.datetime.today() - coh['HIRE_DATE']) / pd.Timedelta(1, 'Y')
```

```python
fig, ax = plt.subplots()
ax.scatter('EXPERIENCE', 'BASE_SALARY', data=coh)
```

```python
import statsmodels.formula.api as smf

# Fit regression model (using the natural log of one of the regressors)
results = smf.ols('BASE_SALARY ~ EXPERIENCE', data=coh).fit()

# Inspect the results
results.summary()
```

```python
results.params
```

```python
fig, ax = plt.subplots()
ax.scatter('EXPERIENCE', 'BASE_SALARY', data=coh)
x = np.arange(0,50,10)
y = results.params.loc['Intercept'] + x * results.params.loc['EXPERIENCE']
ax.plot(x, y, lw=4)
ax.set_ylim(20000)
ax.set_xlim(0, 50)
```

### Problem 7
<span  style="color:green; font-size:16px">Use the **`pd.cut`** function to create three categories of years of experience: 0-10, 10-20 and 20+ and make a pie chart.</span>

```python
categories = pd.cut(coh.EXPERIENCE, bins=[0,10,20,200], labels=['Novice', 'Experienced', 'Senior'])
```

```python
vc = categories.value_counts()
vc = vc / vc.sum()
vc
```

```python
plt.pie(vc, labels=vc.index.values, autopct='%1.1f%%')
plt.legend()
```

### Problem 3
<span  style="color:green; font-size:16px">Use city of Houston data. Rename Black or African American to 'Black' and filter race for only 'Black' or 'White'. Look up how to do a factor plot in seaborn and create a new graph for each POSITION TITLE that has more than 50 occurrences. Then find the mean salary between male and female for every race. Use a boxplot. Scroll down to the [bottom of this page to see an example](http://seaborn.pydata.org/tutorial/categorical.html). </span>

```python
coh = pd.read_csv('data/coh_employee.csv')
```

```python
coh['RACE'] = coh.RACE.replace({'Black or African American':'Black'})
```

```python
coh1 = coh[coh.RACE.isin(['Black', 'White'])]
```

```python
coh1 = coh1.groupby('POSITION_TITLE').filter(lambda x: len(x) > 50)
```

```python
sns.factorplot(x='RACE', y='BASE_SALARY', hue='GENDER', data=coh1, kind='box', col='POSITION_TITLE', sharey=False)
```
